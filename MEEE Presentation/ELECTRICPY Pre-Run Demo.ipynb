```python
# Import Libraries Necessary for Demonstrations
import electricpy as ep
import numpy as np
import matplotlib.pyplot as plt
```

```python
## COMPLEX NUMBERS ##

# Demonstrate Complex Numbers in Python
my_complex = 5 + 15j # Note that Python Natively uses 'j' as the Imaginary Number
print("Native Python Complex Representation:",my_complex)

# Print again using electricpy
ep.cprint(my_complex,label="ELECTRICPY Complex Representation:")

# Plot on Phasor Diagram
ep.phasorplot(my_complex)
```

```python
## POWER TRIANGLE ##

# Demonstrate Plotting a Simple Power Triangle
ep.powertriangle(P=5,S=7,printval=True)
```

```python
## HARMONICS ##

# Create a Harmonic System
my_signal = ep.harmonics([1,0,0.5,0,0.1])
time = np.linspace(0,2*np.pi/60,1000)
# Plot
plt.plot(my_signal(time))
plt.show()

# Demonstrate an FFT of the Signal
DC, A, B = ep.funcfft(my_signal)
ep.fftplot(DC,A,B,title="FFT Plot")
```
