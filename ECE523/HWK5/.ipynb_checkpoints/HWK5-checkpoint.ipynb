```python
# Import Libraries
import numpy as np
import matplotlib.pyplot as plt
import electricpy as ep
from electricpy.constants import *
from tabulate import tabulate as table
```

```python
# Define Parameters
GMR = 0.0244  # feet
dia = 0.721   # inches
Rac = 0.278   # ohm/mi
tmp = 25      # °C
frq = 60      # Hz

# Calculate Rd and Rself
Rd = carson_r * frq
Rself = Rac + Rd
print("Rd:",Rd,"Ω\tRself:",Rself,"Ω")

# Calculate Dab, Dbc, and Dca
Dab = np.sqrt((11-6.5)**2+(4+5)**2)
Dbc = np.sqrt((5+6)**2+(8+6.5-11)**2)
Dca = np.sqrt((6-4)**2+(8)**2)
print("Dab:",Dab,"\tDbc:",Dbc,"\tDca:",Dca)

# Calculate Zperlength Matrix
De = ep.de_calc('Avg')
Zperlen = ep.zperlength(Rd=Rd,Rself=Rself,De=De,Dab=Dab,Dbc=Dbc,Dca=Dca,Ds=GMR)
print("\nZ-per-unit Matrix:")
print(table(np.asarray(np.around(Zperlen,6),dtype=str),tablefmt="fancy_grid"))

# Calculate Sequence Impedances
print("\nSequence Impedance Matrix:")
print(table(np.asarray(np.around(ep.sequencez(Zperlen,resolve=True,round=5),6),dtype=str),tablefmt="fancy_grid"))
print("\nSequence Impedances:")
Z0, Z1, Z2 = ep.sequencez(Zperlen,resolve=True,diag=True,round=5)
print("Z1:",Z1,"Ω","\tZ2:",Z2,"Ω")
```
