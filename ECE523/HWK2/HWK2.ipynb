## Problem 1:
 Determine the symmetrical components for the following currents and sketch a phasor
diagram, based on

a) Phase "a" referenced components

b) Phase "b" referenced components

c) Phase "c" referenced components

## Problem 4: (text 2.10)

```python
# Create Arbitrary Impedance (Load)
Zaa = 0.5+0.1j
Zbb = 0.74+0.15j
Zcc = 0.89+0.11j
Zab = 0.1+0.6j
Zbc = 0.11+0.58j
Zca = 0.12+0.56j
Zabc = np.array([[Zaa,Zab,Zca],
                 [Zab,Zbb,Zbc],
                 [Zca,Zbc,Zcc]])

# Evaluate Sequence Impedance
Z012 = ep.sequencez(Zabc)
print("Sequence Impedance:\n",Z012)

# Create Balanced Voltage Source
Vsrc = ep.phasorlist([[1,0],
                      [1,-120],
                      [1,120]])
V012 = ep.abc_to_seq(Vsrc)
print("\nSequence Voltages of Balanced Voltage Source:")
ep.cprint(V012,label=["V0","V1","V2"])

# Evaluate Sequence Currents
I012 = np.linalg.inv(Z012).dot(V012)
print("\nSequence Currents:")
ep.cprint(I012,label=["I0","I1","I2"])

# Evaluate Phase Currents
Iabc = ep.seq_to_abc(I012)
print("\nPhase Currents:")
ep.cprint(Iabc,label=["IA","IB","IC"])

# Determine Neutral Current
In = np.sum(Iabc)
print("\nNeutral Current:")
ep.cprint(In,label="In:")

# Determine 3*Io
I0_3 = 3*I012[0]
print("\nZero-Sequence Current:")
ep.cprint(I0_3,label="3*I0:")

# Demonstrate that Currents are Equal
ion = r"$$I_{0}-I_{n}="+ep.clatex(I0_3-In,double=True,predollar=False)
```

This demonstrates that $I_n = I_0$ as shown below by their subtracted total:
{{ion}}

## Problem 5: (text 2.13)

```python
# Define Givens
Vab = ep.phasor(1,30)
Vac = ep.phasor(1,-30)
Vbc = ep.phasor(1,-90)
Vca = Vac * 1j**2

# Determine Line-to-Neutral Voltages
Van = ep.phaseline(VLL=Vab,complex=True)
Vbn = ep.phaseline(VLL=Vbc,complex=True)
Vcn = ep.phaseline(VLL=Vca,complex=True)

# Generate Sequence Voltages
V012 = ep.abc_to_seq([Van,Vbn,Vcn])
ep.cprint(V012,label=["V0","V1","V2"])

# Create LaTeX String
Vtex = r'$$V_{012}='+ep.clatex(V012,predollar=False,double=True)
```

## Problem 6:

```python
# Define Givens
CTR_H = 500
CTR_L = 5
CTR = CTR_H/CTR_L
PHS_L = [[12,-87],
        [4,-120],
        [4,120]]
PHS_L = ep.phasorlist(PHS_L)

# a) Primary Currents
print("a) Primary Currents:")
PHS_H = ep.primary(PHS_L,CTR)
ep.cprint(PHS_H)

# b) Secondary Sequence Currents
print("\nb) Sequence Components:")
PHS_L_012 = ep.abc_to_seq(PHS_L)
ep.cprint(PHS_L_012,label=["I0","I1","I2"])

# c) Residual Current
print("\nc) Neutral Current:")
I_r = np.sum(PHS_L)
ep.cprint(I_r,label="Residual:")
ep.cprint(3*PHS_L_012[0],label="3*I0:\t ")

# d) Delta Currents
PHS_L_D = ep.phaseline(Iline=ep.secondary(PHS_L,CTR),complex=True)
PHS_L_D_012 = ep.abc_to_seq(PHS_L_D)
PHS_L_D_012[0] = 0 # Force Zero-Sequence to Zero
print("\nd) Sequence Currents:")
ep.cprint(PHS_L_D_012,label=["I0","I1","I2"])
```
